```python
import numpy as np
import matplotlib.pyplot as plt

def fourier_triangle(x, n):
    result = np.zeros_like(x)
    for k in range(1, n + 1):
        if k % 2 == 1:
            sign = (-1)**((k-1)//2)
            result += sign * (8 / (np.pi**2 * k**2)) * np.sin(k * x)
    return result

x = np.linspace(0, 4*np.pi, 1000)

plt.figure(figsize=(10, 6))
for n in [1, 3, 5, 10, 20]:
    y = fourier_triangle(x, n)
    plt.plot(x, y, label=f'n={n}')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def fourier_square(x, n):
    result = 0.5 * np.ones_like(x)
    for k in range(1, n + 1):
        if k % 2 == 1:
            result += (2 / (k * np.pi)) * np.sin(k * x)
    return result

x = np.linspace(0, 4*np.pi, 1000)

plt.figure(figsize=(10, 6))
for n in [1, 1000000]:
    y = fourier_square(x, n)
    plt.plot(x, y, label=f'n={n}')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
